```python
data_pipeline=api.search_tweets(q='Kazakhstan',count=100,lang='en')
data_pipeline
```

```python
columns = set()
allowed_types = [str, int]
tweets_data = []
for status in data_pipeline:
    print(status.author.screen_name)
    # print(type(vars(status)))
    status_dict = dict(vars(status))
    keys = status_dict.keys()
    single_tweet_data = {"user": status.user.screen_name, "author": status.author.screen_name}
    for k in keys:
        try:
            v_type = type(status_dict[k])
        except:
            v_type = None
        if v_type != None:
            if v_type in allowed_types:
                single_tweet_data[k] = status_dict[k]
                columns.add(k)
    tweets_data.append(single_tweet_data)

        
header_cols = list(columns)
header_cols.append("user")
header_cols.append('author')
```

# The second step-preprocessing

```python
df = pd.DataFrame(tweets_data, columns=header_cols)
df.head()
```

```python
data=df.drop(columns=['quoted_status_id_str','in_reply_to_user_id_str','in_reply_to_screen_name','in_reply_to_user_id','source','id','quoted_status_id','id_str','in_reply_to_status_id','in_reply_to_status_id_str','user','author','source_url'])
data
```

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data['text_without_punct'] = data['text'].apply(lambda x: remove_punct(x))

data.head()
```

```python
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

data['text_tokenized'] = data['text_without_punct'].apply(lambda x: tokenize(x.lower()))

data.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text
```

```python
data['text_nostop'] = data['text_tokenized'].apply(lambda x: remove_stopwords(x))
data.head()
```

```python
wn = nltk.WordNetLemmatizer()
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data['text_lemmatized'] = data['text_nostop'].apply(lambda x: lemmatizing(x))

data.head(10)
```

```python
data.drop(columns=['text_without_punct','text_tokenized','text_nostop'],inplace=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score
```

```python
data['Sentence']=data['text_lemmatized'].apply(lambda x:' '.join(x))
```

```python
data['analysisdict']=data['Sentence'].apply(lambda x:sentiment_analyzer_scores(x))
```
